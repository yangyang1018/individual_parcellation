```python
import nibabel as nib
import numpy as np
import os

Struct_file_dir_path = r'/home/yxl/data/100610/100610'
#输出该路径下所有文件
import os
def list_files_in_directory(directory):
    # 获取目录下所有文件和子目录
    files = os.listdir(directory)
    # 遍历文件和子目录
    for file in files:
        # 拼接完整路径
        file_path = os.path.join(directory, file)
        # 判断是否为文件
        if os.path.isfile(file_path):
            print(file_path)
        # 如果是目录，则递归调用
        elif os.path.isdir(file_path):
            list_files_in_directory(file_path)
            
S_file_list = list_files_in_directory(Struct_file_dir_path)
print(S_file_list)
```

```python
file_dir_root_path = r'/home/yxl/data/100610/100610_3T_rfMRI_REST2_preproc/100610/MNINonLinear/Results/rfMRI_REST2_LR'
#获取该文件下的所有文件目录，并做成一个列表的形式
def list_files_in_directory(directory):
    files = []
    for root, dirs, filenames in os.walk(directory):
        for filename in filenames:
            files.append(os.path.join(root, filename))
    return files
#输出以.nii结尾文件的数据格式

files = list_files_in_directory(file_dir_root_path)
# 读取每个.nii文件并输出其shape和文件名
files = [file for file in files if file.endswith('.nii')]
for file in files:
    if file.endswith('.nii.gz'):
        continue
    try:
        img = nib.load(file)
        print(f"Shape: {img.shape},  File: {os.path.basename(file)}")
    except Exception as e:
        print(f"Error loading file {os.path.basename(file)}: {e}")

```

```python
# 
import os
import nibabel as nib
import numpy as np
func_file = r"F:\hcp_resample\all_subject\100206\rfMRI_REST1_LR_Atlas_hp2000_clean.L.3k_fsavg_L.func.gii"
gii = nib.load(str(func_file))

# 读取所有时间点的数据
n_timepoints = len(gii.darrays)
n_vertices = gii.darrays[0].data.shape[0]
timeseries = np.zeros((n_timepoints, n_vertices))

for t,darray in enumerate(gii.darrays):
    timeseries[t, :] = darray.data

print(f"Shape of timeseries: {timeseries.shape}")

timeseries2 = np.array([darray.data for darray in gii.darrays])

print(f"Shape of timeseries2: {timeseries2.shape}")
# 检查timeseries和timeseries2是否相等
if np.array_equal(timeseries, timeseries2):
    print("timeseries and timeseries2 are equal.")  
else:
    print("timeseries and timeseries2 are not equal.")
```
